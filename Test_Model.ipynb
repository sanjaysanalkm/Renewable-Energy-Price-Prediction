### Importing Packages

### Forecast for Each Site

```python
# Load the model from the file
final_model = joblib.load(model_path+'R_model.pkl')
ghi_forecast_data = pd.read_csv(test_path+'Final_GHI_'+date+'.csv')
ghi_input_data =  ghi_forecast_data['Rajasthan'].values.reshape(-1,1)
gen_forecast = final_model.predict(ghi_input_data)
df = pd.DataFrame(gen_forecast)
writer = pd.ExcelWriter(forecast_path+'Rajasthan_gen_Forecast_'+date+'.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='Forecast')
writer.close()
```

### Merge GEN File

```python
dataframes= pd.read_excel(forecast_input_path+'Rajasthan_gen_Forecast_'+date+'.xlsx')

dataframe = dataframes
# adding column name to the respective columns
dataframe.columns =['S1','Rajasthan']
#drop unused columns
df = dataframe.drop(labels=['S1'], axis = 1, inplace = False)

#SAVE all GEN file to one file
writer = pd.ExcelWriter(output_path+'Final_GEN_'+date+'.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='Generation')
writer.close()
```

### Generate R1 File

```python
dataframe= pd.read_excel(in_path+'Final_GEN_'+date+'.xlsx')
df1 =  dataframe['Rajasthan'].values[:-46]
df = pd.DataFrame(df1)
wb = load_workbook(out_path+'Rajasthan_Forecast_'+old_date+'.xlsx')
ws = wb['WorkSheet 1']
for index, row in df.iterrows():
    cell = 'D%d'  % (index + 27)
    ws[cell] = row[0]    
wb.save(out_path+date+'/'+'Rajasthan_Forecast_'+next_date+'.xlsx')
```
